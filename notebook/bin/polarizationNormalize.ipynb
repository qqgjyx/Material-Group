```python
def show_means(PATH, l=None):
    name = l or "" 
    # Read the Excel file into a pandas DataFrame
    data = pd.read_excel(PATH)
    # Assuming the first column is 'degrees' and should be used as the index
    data.set_index('X', inplace=True)
    # Remove the baseline (first row)
    data = data.iloc[1:, :]
    # Calculate the mean intensity for each spectrum (row)
    mean_intensities = data.mean(axis=1)
    # Plotting the mean intensities
    plt.figure(figsize=(10, 6))
    plt.plot(mean_intensities.index, mean_intensities, marker='o', linestyle='-', color='b')
    plt.xlabel('Degrees')
    plt.ylabel('Mean Intensity')
    plt.title('Mean Intensity of Raman Spectra vs. Degree of Polarization'+name)
    plt.grid(True)
    plt.show()
```

```python
show_means(PATH1,"\nMar 28 NaBaYb(BO$_3$)$_2$ modified")
```

```python
show_means(PATH2, "\nMar 28 NaBaYb(BO$_3$)$_2$")
```

```python
import pandas as pd

# Load the data
file_path = 'vibrational_modes.csv'
df_vibrational_modes = pd.read_csv(file_path)

# Categorize the modes by symmetry
symmetry_counts = df_vibrational_modes['Symmetry'].value_counts().reset_index()
symmetry_counts.columns = ['Symmetry', 'Count']

# Map the symmetry types to the required format
symmetry_map = {
    'A': 'A1g',
    'E': 'Eg',
    'E*': 'Eu'
}

symmetry_counts['Formatted Symmetry'] = symmetry_counts['Symmetry'].map(symmetry_map)

# Summarize the data
total_modes = df_vibrational_modes.shape[0]
raman_active_modes = symmetry_counts[symmetry_counts['Symmetry'].isin(['A', 'E'])]['Count'].sum()

# Format the results
summary = f"""
The polarization-dependent Raman spectra at room temperature under ambient pressure are depicted in Figure 1(c). 
The bulk unit cell comprises a total of {total_modes} normal phonon modes at the Γ point. 
The vibrational modes at the Γ point can be categorized into:
"""

for index, row in symmetry_counts.iterrows():
    summary += f"{row['Count']} {row['Formatted Symmetry']} + "

summary = summary.rstrip(' + ') + ". "
summary += f"""
Among these vibrational modes, {raman_active_modes} are Raman active modes corresponding to the non-degenerate A1g symmetry modes and the doubly degenerate Eg symmetry modes. 
"""

print(summary)
```
